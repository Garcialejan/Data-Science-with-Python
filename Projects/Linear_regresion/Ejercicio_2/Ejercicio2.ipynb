##Reestructuramos el dataframe

```python
df.set_index("id", inplace=True)
df.head()
```

```python
df.columns
```

```python
nuevo_orden_columnas = ['neighbourhood_group', 'neighbourhood', 'room_type', 'minimum_nights',
                        'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365', 'price']
```

```python
df = df.reindex(columns = nuevo_orden_columnas)
df.head()
```

```python
df.rename(columns={'availability_365':'avilability_%'}, inplace=True)
```

```python
df['avilability_%'] = (df['avilability_%']/365)*100
```

```python
df['avilability_%'] = df['avilability_%'].round(2)
```

```python
df.head()
```

- Ya tenemos nuestro data set listo para aplicar las técnicas de **Encoding** y **Scaling**

##Encoding de los registros del tipo string


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data1 = np.array(df['neighbourhood_group'])
```

```python
data1
```

```python
data1.shape
```

```python
le.fit(data1)
```

```python
le.transform(data1)
```

```python
df['neighbourhood_group'] = le.fit_transform(df['neighbourhood_group']) # Unificamos la codificación en una sola linea
```

```python
df.head()
```

- Codificamos el resto de columnas que lo necesitan

```python
df['neighbourhood'] = le.fit_transform(df['neighbourhood'])
```

```python
df['room_type'] = le.fit_transform(df['room_type'])
```

```python
df.head()
```

##Scaling de los datos del tipo numérico


```python
from sklearn.preprocessing import StandardScaler

s_sc = StandardScaler()

df[['minimum_nights',	'number_of_reviews', 'reviews_per_month',	'calculated_host_listings_count',	'avilability_%']] = s_sc.fit_transform(df[['minimum_nights',	'number_of_reviews', 'reviews_per_month',	'calculated_host_listings_count',	'avilability_%']])
```
