```python
df = pd.json_normalize(data)
df
```

```python
# I'm going to reorganise the DataFrame to select only the relevant information on which we will operate
df_resume = df[["profile.name", "profile.dob","profile.location.lat", "profile.location.long"]]
df_resume.columns = ["name", "birth", "loc_lat", "loc_long"]
df_resume
```

## 2.Function to calculate the variance of the **lat**

```python
def var_loc_lat(dataframe):
    varianza = np.var(dataframe['loc_lat'])
    return varianza
```

```python
# We call the function to check it
var = var_loc_lat(df_resume)
var
```

## 3.Function to calculate the outlier of **long**

```python
sample = df_resume["loc_long"]
sample
```

```python
Q1 = np.percentile(sample, 25)
Q3 = np.percentile(sample, 75)
median = np.median(sample)
IQR = Q3 - Q1
print(Q1, Q3, IQR)
print(median)
```

### Now we calculate the outliers

![](https://miro.medium.com/max/9000/1*2c21SkzJMf3frPXPAR_gZA.png)

```python
outlier_inf = (sample < (Q1 - 1.5*(IQR)))
outlier_sup = (sample > (Q3 + 1.5*(IQR)))
df_resume["outlier_inf"] = outlier_inf
df_resume["outlier_sup"] = outlier_sup
```

# SOLUCIÓN

## 1.Calculamos la varianza de las latitudes y filtramos los valores

```python
df_resume["birth"] = pd.to_datetime(df_resume["birth"])
df_resume.info()
```

```python
var = var_loc_lat(df_resume)

# Verificar si la varianza coincide con el día de la fecha en "birth"
df_filtrado_var = df_resume[df_resume['birth'].dt.day == int(str(var)[:2])]
df_filtrado_var
```

## 2.Comprobamos los outliers

```python
df_filtrado_final = df_filtrado_var[(df_filtrado_var['outlier_inf'] == True) | (df_filtrado_var['outlier_sup'] == True)]

df_filtrado_final
```

## 3.Extraemos toda la información de la persona que cumple ambas condiciones del df original

```python
indice_asesino = df_filtrado_final.index[0]

df_asesino = pd.DataFrame(df.values[indice_asesino].reshape(1, -1), columns=df.columns)
```

```python
df_asesino = df_asesino.set_index("profile.name")
df_asesino
```
