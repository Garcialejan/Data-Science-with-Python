# 1. Técnica del With Stament

> - Con la función `open()` creamos una conexión a un archivo, al cual podemos pasar como segundo argumento el tipo de permiso (lectura/**r**ead, escritura/**w**rite, ...)
> - Podemos asignar esta función a una variable y luego popular este archivo (escribir sobre el archivo)

````python
f = open("prueba.txt", "w")

f.write("Mastermind")

f.close()
````

> - Con la función open nosotros estamos creando una conexión a un archivo. Dicha conexión cada vez que se abre se debe cerrar con `f.close`. Si no cerramos el archivo hay muchas funciones de populación del archivo que no se quedan guardadas

## 1.1 With Stament

> - Es una técnica con la que simplificamos los pasos anteriores la cual nos ofrece ciertas convenencias
> - La sintaxis es la siguiente:
>> Keyworth `with ... as variable:` para juntar los pasos en solo dos líneas de código

````python
with open("prueba.txt", "w") as f:

  f.write("MMASTERMIND 2")
````

> - Con este método no es necesario cerrar el archivo ya que crea una instancia temporal que se cierra cada vez que se ejecuta. Facilita mucho la gestión de errores.
> - Se puede ejecutar este with stament con archivos JSON y asignar una variable al contenido de dicho archivo JSON.
> - Para leer el contenido de estos archivos tenemos que importar la librería de JSON y utilizar una función que viene integrada. Viene installada directamente, por lo que no es necesaria la instalación.

````python
import json
````

````python
path = "/content/sample_data/anscombe.json"

with open(path, "r") as f:

  data = json.load(f)

data
````

# 2. Transformar archivos JSON en DataFrames

## 2.1 Función `pd.read_json()`
> - Pandas nos ofrece una función muy sencilla para realizar esta transformación sin necesidad de utilizar el with statement
> - Función `pd.read_json()`

````python
import pandas as pd
````

````python
path = "/content/sample_data/anscombe.json"

df = pd.read_json(path)
df
````

> - Pueden haber fallos estructurales en este tipo de archivos JSON cuando los transformamos en DF. Por ejemplo, que en la misma columna aparezca más de una variable. Esto puede darse cuando tengamos un diccionario dentro del diccionario que queremos abrir
> - Función de la librería de json para visualizar el contenido de una manera mejor estructurada. Función `print(json.dumps(archivo))`
> - Podemos pasar un argumento a esta función que permite realizar una indentación, muy típica de python. El argumento es el siguiente: `print(json.dumps(archivo, indent = 4))`. Esto indica un espaciamiento de 4 caracteres por cadas llave
> - Está función es imprescindible usarlo dentro de un print, ya que en realidad no está pensada para realizar esta previsualización del diccionario. Lo que en realidad hace esta función es guardar un diccionario indentando la separación que nosotros le indiquemos.

````python
path = "/content/sample_data/anscombe.json"

with open(path, "r") as f:

  data = json.load(f)

print(json.dumps(data, indent = 4))
````
